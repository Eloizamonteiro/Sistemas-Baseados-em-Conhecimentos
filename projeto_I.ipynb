```python
from experta import *

# Fato
class Plantacao(Fact):
    pass

# Motor do Sistema

class SBCPlantacao(KnowledgeEngine):

    def __init__(self):
        super().__init__()
        self.diagnostico_principal = []
        self.recomendacoes = []


    # Nutrição (R1–R5)


    @Rule(Plantacao(folhas_amarelas=True))
    def r1(self):
        print("R1 → Suspeita de deficiência nutricional")
        self.declare(Fact(def_nutricional=True))

    @Rule(Fact(def_nutricional=True),
          Plantacao(crescimento_lento=True))
    def r2(self):
        print("R2 → Deficiência de nitrogênio")
        self.declare(Fact(def_nitrogenio=True))

    @Rule(Fact(def_nitrogenio=True))
    def r3(self):
        print("R3 → Aplicar adubação nitrogenada")
        self.diagnostico_principal.append("Deficiência nutricional")
        self.recomendacoes.append("Adubação nitrogenada")

    @Rule(Fact(def_nitrogenio=True))
    def r4(self):
        print("R4 → Plano de correção nutricional criado")
        self.diagnostico_principal.append("Correção nutricional")

    @Rule(Fact(def_nitrogenio=True))
    def r5(self):
        print("R5 → Monitorar recuperação da planta")


    # Irrigação (R6–R10)


    @Rule(Plantacao(solo_seco=True))
    def r6(self):
        print("R6 → Falta de irrigação suspeita")
        self.declare(Fact(falta_agua=True))

    @Rule(Fact(falta_agua=True),
          Plantacao(folhas_murchas=True))
    def r7(self):
        print("R7 → Estresse hídrico")
        self.declare(Fact(estresse_hidrico=True))

    @Rule(Fact(estresse_hidrico=True))
    def r8(self):
        print("R8 → Aumentar irrigação")
        self.diagnostico_principal.append("Estresse hídrico")
        self.recomendacoes.append("Aumentar irrigação")

    @Rule(Fact(estresse_hidrico=True))
    def r9(self):
        print("R9 → Plano de irrigação criado")

    @Rule(Fact(estresse_hidrico=True))
    def r10(self):
        print("R10 → Monitorar umidade do solo")


    # Fungos (R11–R15)

    @Rule(Plantacao(manchas_escuras=True))
    def r11(self):
        print("R11 → Suspeita de fungos")
        self.declare(Fact(doenca_fungo=True))

    @Rule(Fact(doenca_fungo=True),
          Plantacao(alta_umidade=True))
    def r12(self):
        print("R12 → Risco elevado de fungos")
        self.declare(Fact(risco_fungo=True))

    @Rule(Fact(risco_fungo=True))
    def r13(self):
        print("R13 → Aplicar fungicida")
        self.diagnostico_principal.append("Doença fúngica")
        self.recomendacoes.append("Fungicida")

    @Rule(Fact(risco_fungo=True))
    def r14(self):
        print("R14 → Monitorar evolução da doença")

    @Rule(Fact(risco_fungo=True),
          NOT(Fact(fungicida_aplicado=True)),
          salience=10)
    def r15(self):
        print("R15 🔥 ALERTA: fungicida não aplicado!")


    # Praga (R16–R20)

    @Rule(Plantacao(folhas_perfuradas=True))
    def r16(self):
        print("R16 → Suspeita de insetos")
        self.declare(Fact(insetos=True))

    @Rule(Fact(insetos=True),
          Plantacao(presenca_lagartas=True))
    def r17(self):
        print("R17 → Infestação por lagartas")
        self.declare(Fact(lagartas=True))

    @Rule(Fact(lagartas=True))
    def r18(self):
        print("R18 → Controle biológico recomendado")
        self.diagnostico_principal.append("Pragas (lagartas)")
        self.recomendacoes.append("Controle biológico")

    @Rule(Fact(lagartas=True),
          Plantacao(dano_severo=True))
    def r19(self):
        print("R19 → Controle químico necessário")
        self.declare(Fact(controle_quimico=True))

    @Rule(Fact(controle_quimico=True),
          salience=5)
    def r20(self):
        print("R20 → ALERTA: controle químico urgente!")
```

```python
def coletar_sintomas():
    print("\nRESPONDA COM S OU N\n")

    def s(msg):
        return input(msg + " (s/n): ").strip().lower() == "s"

    return Plantacao(
        folhas_amarelas=s("Folhas amarelas"),
        crescimento_lento=s("Crescimento lento"),
        solo_seco=s("Solo seco"),
        folhas_murchas=s("Folhas murchas"),
        manchas_escuras=s("Manchas escuras"),
        alta_umidade=s("Alta umidade"),
        folhas_perfuradas=s("Folhas perfuradas"),
        presenca_lagartas=s("Presença de lagartas"),
        dano_severo=s("Dano severo")
    )
```

```python
engine = SBCPlantacao()
engine.reset()

print("SISTEMA ESPECIALISTA AGRÍCOLA")

sintomas = coletar_sintomas()
engine.declare(sintomas)

print("\nPROCESSANDO...\n")
engine.run()


# Diágnóstico

print("\nDIAGNÓSTICO FINAL:")

if engine.diagnostico_principal:
    vistos = []
    for d in engine.diagnostico_principal:
        if d not in vistos:
            vistos.append(d)

    for d in vistos:
        print("-", d)
else:
    print("✔ Nenhum problema crítico detectado")


# Recomendaçõesnnn
print("\nRECOMENDAÇÕES:")

if engine.recomendacoes:
    for r in set(engine.recomendacoes):
        print("-", r)
```
