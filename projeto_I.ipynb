```python
from experta import KnowledgeEngine, Rule, Fact, MATCH, AS
from experta import P    # predicates (e.g. P(lambda x: x > 37.8))
from experta import OR, AND, NOT
```

```python
class Plantacao(Fact):
  # Informações observadas na plantação
  pass
```

```python
# MOTOR DE INFERÊNCIA
class DiagnosticoPlantacao(KnowledgeEngine):


    # NUTRIÇÃO (R1 - R5)

    @Rule(Plantacao(folhas_amarelas=True))
    def r1(self):
        print("R1 -> Folhas amarelas: suspeita de deficiência nutricional")
        self.declare(Fact(def_nutricional=True))

    @Rule(Fact(def_nutricional=True),
          Plantacao(crescimento_lento=True))
    def r2(self):
        print("R2 -> Deficiência nutricional + crescimento lento: falta de nitrogênio")
        self.declare(Fact(def_nitrogenio=True))

    @Rule(Fact(def_nitrogenio=True))
    def r3(self):
        print("R3 -> Diagnóstico: deficiência de nitrogênio")
        self.declare(Fact(adubacao_n=True))

    @Rule(Fact(adubacao_n=True))
    def r4(self):
        print("R4 -> Ação: aplicar adubação nitrogenada")
        self.declare(Fact(plano_nutri=True))

    @Rule(Fact(plano_nutri=True))
    def r5(self):
        print("R5 -> Monitorar recuperação nutricional")


    # IRRIGAÇÃO (R6 - R10)

    @Rule(Plantacao(solo_seco=True))
    def r6(self):
        print("R6 -> Solo seco: suspeita de falta de irrigação")
        self.declare(Fact(falta_agua=True))

    @Rule(Fact(falta_agua=True),
          Plantacao(folhas_murchas=True))
    def r7(self):
        print("R7 -> Falta de água + murcha: estresse hídrico")
        self.declare(Fact(estresse_hidrico=True))

    @Rule(Fact(estresse_hidrico=True))
    def r8(self):
        print("R8 -> Recomendação: aumentar irrigação")
        self.declare(Fact(aumentar_irr=True))

    @Rule(Fact(aumentar_irr=True))
    def r9(self):
        print("R9 -> Criar plano de irrigação")
        self.declare(Fact(plano_irr=True))

    @Rule(Fact(plano_irr=True))
    def r10(self):
        print("R10 -> Monitorar umidade do solo")


    # FUNGOS (R11 - R15)

    @Rule(Plantacao(manchas_escuras=True))
    def r11(self):
        print("R11 -> Manchas escuras: suspeita de fungos")
        self.declare(Fact(doenca_fungo=True))

    @Rule(Fact(doenca_fungo=True),
          Plantacao(alta_umidade=True))
    def r12(self):
        print("R12 -> Umidade alta + fungo: risco elevado")
        self.declare(Fact(risco_fungo=True))

    @Rule(Fact(risco_fungo=True))
    def r13(self):
        print("R13 -> Recomendar fungicida")
        self.declare(Fact(fungicida=True))

    @Rule(Fact(fungicida=True))
    def r14(self):
        print("R14 -> Monitorar evolução da doença")

    # REGRA COM CONFLITO (salience + NOT)
    @Rule(Fact(risco_fungo=True),
          NOT(Fact(fungicida_aplicado=True)),
          salience=10)
    def r15(self):
        print("R15 🔥 ALERTA: risco alto de fungos sem tratamento!")


    # PRAGAS (R16 - R20)

    @Rule(Plantacao(folhas_perfuradas=True))
    def r16(self):
        print("R16 -> Folhas perfuradas: suspeita de insetos")
        self.declare(Fact(insetos=True))

    @Rule(Fact(insetos=True),
          Plantacao(presenca_lagartas=True))
    def r17(self):
        print("R17 -> Infestação por lagartas detectada")
        self.declare(Fact(lagartas=True))

    @Rule(Fact(lagartas=True))
    def r18(self):
        print("R18 -> Recomendação: controle biológico")

    @Rule(Fact(lagartas=True),
          Plantacao(dano_severo=True))
    def r19(self):
        print("R19 -> Dano severo: controle químico necessário")
        self.declare(Fact(controle_quimico=True))

    @Rule(Fact(controle_quimico=True),
          salience=5)
    def r20(self):
        print("R20 -> ALERTA: aplicação química urgente!")
```

```python
engine = DiagnosticoPlantacao()
engine.reset()

```
